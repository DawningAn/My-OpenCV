```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.dpi'] = 150
```

```python
img = cv2.imread('faces1.jpg')
```

```python
img.shape
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
# 构造Haar检测器
face_detector = cv2.CascadeClassifier('./cascades/haarcascade_frontalface_default.xml')
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# 检测结果
detections = face_detector.detectMultiScale(gray, scaleFactor=1.6)
detections  # 人脸坐标
```

```python
for (x, y, w, h) in detections:
    print(x, y, w, h)
```

```python
# 绘制矩形
for (x, y, w, h) in detections:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 3)
```

```python
plt.imshow(img)
```

```python
# 调节参数
img = cv2.imread('faces1.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
detections = face_detector.detectMultiScale(gray, scaleFactor=1.6)
for (x, y, w, h) in detections:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 3)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
# 调节参数
# scaleFactor：调整图片尺寸
# minNeighbors：候选人脸数量
# minSize：最小人脸尺寸
# maxSize：最大人脸尺寸
img1 = cv2.imread('faces2.jpg')
gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
detections = face_detector.detectMultiScale(gray, scaleFactor=1.2, minNeighbors=7,minSize=(10,10),maxSize=(70,70))
for (x, y, w, h) in detections:
    cv2.rectangle(img1, (x, y), (x + w, y + h), (0, 255, 0), 3)
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
```
